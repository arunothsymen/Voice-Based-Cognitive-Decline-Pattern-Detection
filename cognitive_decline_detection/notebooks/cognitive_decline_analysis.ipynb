```python
# Voice-Based Cognitive Decline Pattern Detection
# =============================================
#
# This notebook implements a proof-of-concept pipeline for detecting potential cognitive
# decline patterns using voice data samples. It extracts acoustic and linguistic features
# from speech that may indicate cognitive impairment.

import os
import sys
import numpy as np
import pandas as pd
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.ensemble import IsolationForest
import speech_recognition as sr
import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
import warnings
import wave
import contextlib
import glob
from IPython.display import display, Audio
from tqdm.notebook import tqdm
import time

# Suppress warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)

# Download NLTK resources if needed
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
    
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

# Define paths
BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
DATA_DIR = os.path.join(BASE_DIR, "data")
RAW_AUDIO_DIR = os.path.join(DATA_DIR, "raw_audio")
PROCESSED_AUDIO_DIR = os.path.join(DATA_DIR, "processed_audio")
TRANSCRIPT_DIR = os.path.join(DATA_DIR, "transcripts")

# Create directories if they don't exist
os.makedirs(RAW_AUDIO_DIR, exist_ok=True)
os.makedirs(PROCESSED_AUDIO_DIR, exist_ok=True)
os.makedirs(TRANSCRIPT_DIR, exist_ok=True)

print(f"Project directories setup complete.")
print(f"Raw audio directory: {RAW_AUDIO_DIR}")
print(f"Processed audio directory: {PROCESSED_AUDIO_DIR}")
print(f"Transcript directory: {TRANSCRIPT_DIR}")

# ## 1. Data Collection and Exploration
# 
# First, let's check if we have audio files in our raw audio directory.

audio_files = [f for f in os.listdir(RAW_AUDIO_DIR) if f.endswith(('.wav', '.mp3'))]
print(f"Found {len(audio_files)} audio files in {RAW_AUDIO_DIR}")

if len(audio_files) == 0:
    print("No audio files found! Please add audio files to the raw_audio directory.")
    # You can uncomment this code to use demo files if needed
    # !wget -P {RAW_AUDIO_DIR} https://example.com/sample_audio1.wav
    # !wget -P {RAW_AUDIO_DIR} https://example.com/sample_audio2.wav
    # audio_files = [f for f in os.listdir(RAW_AUDIO_DIR) if f.endswith(('.wav', '.mp3'))]
    # print(f"Now found {len(audio_files)} audio files")
else:
    print("Audio files found:")
    for file in audio_files:
        print(f" - {file}")

# ### 1.1 Audio File Exploration
# 
# Let's examine the properties of our audio files.

def get_audio_file_info(file_path):
    """Get basic information about an audio file."""
    try:
        # Get duration using wave module
        with contextlib.closing(wave.open(file_path, 'r')) as f:
            frames = f.getnframes()
            rate = f.getframerate()
            duration = frames / float(rate)
            channels = f.getnchannels()
            sample_width = f.getsampwidth()
            
        # Get additional info using librosa
        y, sr = librosa.load(file_path, sr=None)
        rms = librosa.feature.rms(y=y)[0].mean()
        
        return {
            'file_name': os.path.basename(file_path),
            'channels': channels,
            'sample_rate': sr,
            'sample_width': sample_width,
            'duration': duration,
            'rms_energy': rms
        }
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

# Get info for each audio file
audio_info = []
for file in audio_files:
    file_path = os.path.join(RAW_AUDIO_DIR, file)
    info = get_audio_file_info(file_path)
    if info:
        audio_info.append(info)

# Create a DataFrame from the audio info
audio_df = pd.DataFrame(audio_info)
if not audio_df.empty:
    display(audio_df)
    
    # Plot duration distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(audio_df['duration'], bins=10)
    plt.title('Audio Duration Distribution')
    plt.xlabel('Duration (seconds)')
    plt.ylabel('Count')
    plt.grid(True, alpha=0.3)
    plt.show()
    
    # Plot sample rate distribution
    plt.figure(figsize=(10, 6))
    sns.countplot(y='sample_rate', data=audio_df)
    plt.title('Audio Sample Rate Distribution')
    plt.xlabel('Count')
    plt.ylabel('Sample Rate (Hz)')
    plt.grid(True, alpha=0.3)
    plt.show()
else:
    print("No valid audio files found or could not process any of the audio files.")

# ### 1.2 Audio Signal Visualization
# 
# Let's visualize the waveform and spectrogram of one of our audio files.

def visualize_audio(file_path):
    """Visualize waveform and spectrogram of an audio file."""
    # Load audio file
    y, sr = librosa.load(file_path, sr=None)
    
    # Plot waveform
    plt.figure(figsize=(15, 10))
    
    plt.subplot(3, 1, 1)
    librosa.display.waveshow(y, sr=sr)
    plt.title('Waveform')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    
    # Plot spectrogram
    plt.subplot(3, 1, 2)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='hz')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram')
    
    # Plot mel spectrogram
    plt.subplot(3, 1, 3)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    librosa.display.specshow(mel_spec_db, sr=sr, x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel Spectrogram')
    
    plt.tight_layout()
    plt.show()
    
    # Play audio if in a Jupyter notebook
    display(Audio(y, rate=sr))

# Visualize the first audio file
if audio_files:
    first_file = os.path.join(RAW_AUDIO_DIR, audio_files[0])
    print(f"Visualizing: {audio_files[0]}")
    visualize_audio(first_file)

# ## 2. Feature Extraction Pipeline
# 
# Now, let's create our feature extraction pipeline to identify potential cognitive decline indicators.

# ### 2.1 Audio Preprocessing and Feature Extraction

def preprocess_audio(file_path, target_sr=22050):
    """
    Preprocess audio file to make it suitable for feature extraction.
    
    Args:
        file_path: Path to the audio file
        target_sr: Target sample rate
    
    Returns:
        y: Audio time series
        sr: Sample rate
    """
    # Load audio
    y, sr = librosa.load(file_path, sr=target_sr)
    
    # Convert to mono if stereo
    if len(y.shape) > 1:
        y = librosa.to_mono(y)
    
    # Normalize audio
    y = librosa.util.normalize(y)
    
    # Trim silence
    y, _ = librosa.effects.trim(y, top_db=20)
    
    return y, sr

def extract_audio_features(file_path):
    """
    Extract acoustic features that may indicate cognitive decline.
    
    Args:
        file_path: Path to the audio file
    
    Returns:
        Dictionary of extracted features
    """
    # Preprocess audio
    y, sr = preprocess_audio(file_path)
    
    # Calculate duration
    duration = librosa.get_duration(y=y, sr=sr)
    
    # Features dictionary
    features = {'file_name': os.path.basename(file_path), 'duration': duration}
    
    # 1. Speech Rate & Rhythm Features
    # ------------------------------
    
    # Onset detection (syllable-like events)
    onset_env = librosa.onset.onset_strength(y=y, sr=sr)
    onset_times = librosa.times_like(onset_env, sr=sr)
    onsets = librosa.onset.onset_detect(onset_envelope=onset_env, sr=sr, 
                                        units='time', pre_max=0.5, post_max=0.5, 
                                        pre_avg=0.5, post_avg=0.5, delta=0.1)
    
    # Estimated syllable count (based on onsets)
    syllable_count = len(onsets)
    
    # Speech rate (syllables per minute)
    if duration > 0:
        speech_rate = syllable_count / duration * 60
    else:
        speech_rate = 0
    features['speech_rate'] = speech_rate
    
    # 2. Pause & Silence Features
    # --------------------------
    
    # Frame-level energy calculation
    frame_length = 2048
    hop_length = 512
    
    # RMS energy in frames
    rms = librosa.feature.rms(y=y, frame_length=frame_length, hop_length=hop_length)[0]
    
    # Silence detection (frames with energy below threshold)
    threshold = 0.01 * np.mean(rms)
    silence_frames = np.where(rms < threshold)[0]
    
    # Convert to time and find continuous silence segments
    silence_times = librosa.frames_to_time(silence_frames, sr=sr, hop_length=hop_length)
    
    # Group continuous silence segments
    if len(silence_times) > 0:
        silence_gaps = np.split(silence_times, np.where(np.diff(silence_times) > 0.1)[0] + 1)
        
        # Calculate pause durations
        pause_durations = []
        for gap in silence_gaps:
            if len(gap) > 0:
                pause_duration = gap[-1] - gap[0]
                # Only count pauses longer than 0.15 seconds
                if pause_duration > 0.15:
                    pause_durations.append(pause_duration)
        
        # Calculate pause features
        if pause_durations:
            features['pause_count'] = len(pause_durations)
            features['avg_pause_duration'] = np.mean(pause_durations)
            features['max_pause_duration'] = np.max(pause_durations)
            features['pause_rate'] = len(pause_durations) / duration if duration > 0 else 0
        else:
            features['pause_count'] = 0
            features['avg_pause_duration'] = 0
            features['max_pause_duration'] = 0
            features['pause_rate'] = 0
    else:
        features['pause_count'] = 0
        features['avg_pause_duration'] = 0
        features['max_pause_duration'] = 0
        features['pause_rate'] = 0
    
    # 3. Voice Quality Features
    # ------------------------
    
    # Pitch (fundamental frequency) analysis
    try:
        # F0 contour extraction using pYIN
        f0, voiced_flag, voiced_probs = librosa.pyin(y, 
                                                  fmin=librosa.note_to_hz('C2'), 
                                                  fmax=librosa.note_to_hz('C7'),
                                                  sr=sr, frame_length=2048, 
                                                  hop_length=512)
        
        # Clean up NaN values
        f0_clean = f0[~np.isnan(f0)]
        
        if len(f0_clean) > 0:
            # Pitch statistics
            features['pitch_mean'] = np.mean(f0_clean)
            features['pitch_std'] = np.std(f0_clean)
            features['pitch_min'] = np.min(f0_clean)
            features['pitch_max'] = np.max(f0_clean)
            features['pitch_range'] = features['pitch_max'] - features['pitch_min']
            
            # Pitch variability (coefficient of variation)
            if features['pitch_mean'] > 0:
                features['pitch_cv'] = features['pitch_std'] / features['pitch_mean']
            else:
                features['pitch_cv'] = 0
            
            # Jitter (cycle-to-cycle pitch variation)
            if len(f0_clean) > 1:
                pitch_diffs = np.abs(np.diff(f0_clean))
                jitter = np.mean(pitch_diffs) / np.mean(f0_clean) if np.mean(f0_clean) > 0 else 0
                features['jitter'] = jitter
            else:
                features['jitter'] = 0
        else:
            features['pitch_mean'] = 0
            features['pitch_std'] = 0
            features['pitch_min'] = 0
            features['pitch_max'] = 0
            features['pitch_range'] = 0
            features['pitch_cv'] = 0
            features['jitter'] = 0
    except Exception as e:
        print(f"Error extracting pitch features: {e}")
        features['pitch_mean'] = 0
        features['pitch_std'] = 0
        features['pitch_min'] = 0
        features['pitch_max'] = 0
        features['pitch_range'] = 0
        features['pitch_cv'] = 0
        features['jitter'] = 0
    
    # 4. Spectral Features
    # -------------------
    
    # Spectral centroid (weighted mean of frequencies)
    spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)[0]
    features['spectral_centroid_mean'] = np.mean(spectral_centroid)
    features['spectral_centroid_std'] = np.std(spectral_centroid)
    
    # Spectral rolloff (frequency below which 85% of energy is contained)
    spectral_rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)[0]
    features['spectral_rolloff_mean'] = np.mean(spectral_rolloff)
    features['spectral_rolloff_std'] = np.std(spectral_rolloff)
    
    # Spectral contrast (difference between peaks and valleys)
    spectral_contrast = librosa.feature.spectral_contrast(y=y, sr=sr)
    features['spectral_contrast_mean'] = np.mean(spectral_contrast)
    
    # MFCCs (for timbre and voice quality)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    for i in range(13):
        features[f'mfcc{i+1}_mean'] = np.mean(mfccs[i])
        features[f'mfcc{i+1}_std'] = np.std(mfccs[i])
    
    # 5. Speech Rhythm Features
    # -----------------------
    
    # Tempo estimation
    try:
        tempo, _ = librosa.beat.beat_track(y=y, sr=sr)
        features['tempo'] = tempo
    except:
        features['tempo'] = 0
    
    # Energy distribution over time
    rms_std = np.std(rms) if len(rms) > 0 else 0
    features['energy_std'] = rms_std
    
    return features

def extract_audio_features_batch(audio_files):
    """
    Extract acoustic features for a batch of audio files.
    
    Args:
        audio_files: List of audio file paths
    
    Returns:
        DataFrame of extracted features
    """
    features_list = []
    
    for file in tqdm(audio_files, desc="Extracting audio features"):
        file_path = os.path.join(RAW_AUDIO_DIR, file)
        try:
            features = extract_audio_features(file_path)
            features_list.append(features)
        except Exception as e:
            print(f"Error processing {file}: {e}")
    
    return pd.DataFrame(features_list)

# ### 2.2 Speech-to-Text Transcription
# 
# Let's transcribe the audio files to extract linguistic features.

def transcribe_audio(file_path):
    """
    Transcribe an audio file to text.
    
    Args:
        file_path: Path to the audio file
    
    Returns:
        Transcribed text
    """
    # Check if transcript already exists
    transcript_file = os.path.join(TRANSCRIPT_DIR, os.path.basename(file_path).replace('.wav', '.txt'))
    
    if os.path.exists(transcript_file):
        with open(transcript_file, 'r') as f:
            return f.read().strip()
    
    # Initialize recognizer
    recognizer = sr.Recognizer()
    
    try:
        # Load audio file
        with sr.AudioFile(file_path) as source:
            audio_data = recognizer.record(source)
            
            # Use speech recognition to transcribe
            text = recognizer.recognize_google(audio_data)
            
            # Save transcript
            with open(transcript_file, 'w') as f:
                f.write(text)
            
            return text
    except Exception as e:
        print(f"Error transcribing {file_path}: {e}")
        return ""

def transcribe_audio_batch(audio_files):
    """
    Transcribe a batch of audio files to text.
    
    Args:
        audio_files: List of audio file paths
    
    Returns:
        Dictionary mapping file names to transcripts
    """
    transcripts = {}
    
    for file in tqdm(audio_files, desc="Transcribing audio"):
        file_path = os.path.join(RAW_AUDIO_DIR, file)
        try:
            transcript = transcribe_audio(file_path)
            transcripts[file] = transcript
        except Exception as e:
            print(f"Error transcribing {file}: {e}")
            transcripts[file] = ""
    
    return transcripts

# ### 2.3 NLP Feature Extraction
# 
# Let's extract linguistic features from the transcripts that may indicate cognitive decline.

def extract_nlp_features(transcript, file_name):
    """
    Extract linguistic features from a transcript that may indicate cognitive decline.
    
    Args:
        transcript: Transcribed text
        file_name: Name of the audio file
    
    Returns:
        Dictionary of extracted features
    """
    features = {'file_name': file_name}
    
    # If transcript is empty, return empty features
    if not transcript or transcript.strip() == "":
        features['word_count'] = 0
        features['unique_words'] = 0
        features['lexical_diversity'] = 0
        features['avg_words_per_sentence'] = 0
        features['hesitation_count'] = 0
        features['hesitation_rate'] = 0
        features['repetition_count'] = 0
        features['repetition_rate'] = 0
        features['sentence_count'] = 0
        return features
    
    # Tokenize text
    tokens = word_tokenize(transcript.lower())
    sentences = sent_tokenize(transcript)
    
    # 1. Basic Counts
    # --------------
    word_count = len(tokens)
    unique_words = len(set(tokens))
    
    features['word_count'] = word_count
    features['unique_words'] = unique_words
    features['sentence_count'] = len(sentences)
    
    # 2. Lexical Diversity
    # ------------------
    if word_count > 0:
        features['lexical_diversity'] = unique_words / word_count
    else:
        features['lexical_diversity'] = 0
    
    # 3. Hesitation Markers
    # -------------------
    hesitation_patterns = r'\b(um|uh|hmm|err|like|you know)\b'
    hesitation_count = len(re.findall(hesitation_patterns, transcript.lower()))
    
    features['hesitation_count'] = hesitation_count
    if word_count > 0:
        features['hesitation_rate'] = hesitation_count / word_count
    else:
        features['hesitation_rate'] = 0
    
    # 4. Word Repetitions
    # -----------------
    # Look for immediate word repetitions (e.g., "the the", "I I")
    repetition_count = 0
    for i in range(len(tokens) - 1):
        if tokens[i] == tokens[i+1]:
            repetition_count += 1
    
    features['repetition_count'] = repetition_count
    if word_count > 0:
        features['repetition_rate'] = repetition_count / word_count
    else:
        features['repetition_rate'] = 0
    
    # 5. Sentence Complexity
    # --------------------
    sentence_lengths = [len(word_tokenize(s)) for s in sentences]
    if sentence_lengths:
        features['avg_words_per_sentence'] = np.mean(sentence_lengths)
        features['min_words_per_sentence'] = np.min(sentence_lengths)
        features['max_words_per_sentence'] = np.max(sentence_lengths)
        features['std_words_per_sentence'] = np.std(sentence_lengths)
    else:
        features['avg_words_per_sentence'] = 0
        features['min_words_per_sentence'] = 0
        features['max_words_per_sentence'] = 0
        features['std_words_per_sentence'] = 0
    
    # 6. Content Analysis
    # ----------------
    # Filler words
    filler_patterns = r'\b(well|so|actually|basically|literally|anyway|whatever|kind of|sort of)\b'
    filler_count = len(re.findall(filler_patterns, transcript.lower()))
    
    features['filler_count'] = filler_count
    if word_count > 0:
        features['filler_rate'] = filler_count / word_count
    else:
        features['filler_rate'] = 0
    
    # 7. Incomplete Sentences
    # ---------------------
    # Simple heuristic: sentences that don't end with proper punctuation
    proper_ending_pattern = r'[.!?]$'
    incomplete_sentences = sum(1 for s in sentences if not re.search(proper_ending_pattern, s.strip()))
    
    features['incomplete_sentences'] = incomplete_sentences
    if len(sentences) > 0:
        features['incomplete_sentence_rate'] = incomplete_sentences / len(sentences)
    else:
        features['incomplete_sentence_rate'] = 0
    
    return features

def extract_nlp_features_batch(transcripts):
    """
    Extract linguistic features for a batch of transcripts.
    
    Args:
        transcripts: Dictionary mapping file names to transcripts
    
    Returns:
        DataFrame of extracted features
    """
    features_list = []
    
    for file_name, transcript in tqdm(transcripts.items(), desc="Extracting NLP features"):
        try:
            features = extract_nlp_features(transcript, file_name)
            features_list.append(features)
        except Exception as e:
            print(f"Error processing transcript for {file_name}: {e}")
    
    return pd.DataFrame(features_list)

# ## 3. Run Feature Extraction Pipeline
# 
# Now let's run our feature extraction pipeline on the audio files.

print("Starting feature extraction pipeline...")

# Extract audio features
print("Extracting audio features...")
audio_features_df = extract_audio_features_batch([f for f in audio_files if f.endswith('.wav')])

# Transcribe audio files
print("Transcribing audio files...")
transcripts = transcribe_audio_batch([f for f in audio_files if f.endswith('.wav')])

# Extract NLP features
print("Extracting NLP features...")
nlp_features_df = extract_nlp_features_batch(transcripts)

# Merge features
print("Merging features...")
features_df = pd.merge(audio_features_df, nlp_features_df, on='file_name', how='outer')

# Display the resulting features
print("\nExtracted features:")
display(features_df)

# Save the features to CSV
features_csv_path = os.path.join(DATA_DIR, "features.csv")
features_df.to_csv(features_csv_path, index=False)
print(f"Features saved to {features_csv_path}")

# ## 4. Data Visualization
# 
# Let's visualize key features to gain insights.

# Check if we have valid features
if not features_df.empty:
    # Remove any rows with NaN values
    features_df = features_df.dropna()
    
    # Audio features visualization
    plt.figure(figsize=(15, 10))
    
    # Speech Rate
    plt.subplot(2, 2, 1)
    sns.barplot(x='file_name', y='speech_rate', data=features_df)
    plt.title('Speech Rate (syllables/min)')
    plt.xticks(rotation=45)
    
    # Pause Rate
    plt.subplot(2, 2, 2)
    sns.barplot(x='file_name', y='pause_rate', data=features_df)
    plt.title('Pause Rate (pauses/sec)')
    plt.xticks(rotation=45)
    
    # Pitch Variability
    plt.subplot(2, 2, 3)
    if 'pitch_cv' in features_df.columns:
        sns.barplot(x='file_name', y='pitch_cv', data=features_df)
        plt.title('Pitch Coefficient of Variation')
    else:
        plt.title('Pitch Variability - Not Available')
    plt.xticks(rotation=45)
    
    # Jitter
    plt.subplot(2, 2, 4)
    if 'jitter' in features_df.columns:
        sns.barplot(x='file_name', y='jitter', data=features_df)
        plt.title('Jitter (voice instability)')
    else:
        plt.title('Jitter - Not Available')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    # NLP features visualization
    plt.figure(figsize=(15, 10))
    
    # Lexical Diversity
    plt.subplot(2, 2, 1)
    sns.barplot(x='file_name', y='lexical_diversity', data=features_df)
    plt.title('Lexical Diversity')
    plt.xticks(rotation=45)
    
    # Hesitation Rate
    plt.subplot(2, 2, 2)
    sns.barplot(x='file_name', y='hesitation_rate', data=features_df)
    plt.title('Hesitation Rate')
    plt.xticks(rotation=45)
    
    # Repetition Rate
    plt.subplot(2, 2, 3)
    sns.barplot(x='file_name', y='repetition_rate', data=features_df)
    plt.title('Word Repetition Rate')
    plt.xticks(rotation=45)
    
    # Average Words per Sentence
    plt.subplot(2, 2, 4)
    sns.barplot(x='file_name', y='avg_words_per_sentence', data=features_df)
    plt.title('Avg Words per Sentence')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    # Feature correlation heatmap
    plt.figure(figsize=(16, 12))
    
    # Select numeric columns
    numeric_cols = features_df.select_dtypes(include=[np.number]).columns
    
    # Remove columns with zero variance
    var = features_df[numeric_cols].var()
    numeric_cols = var[var > 0].index.tolist()
    
    # Calculate correlation matrix
    corr = features_df[numeric_cols].corr()
    
# Continue from the existing code

# Feature correlation heatmap
plt.figure(figsize=(16, 12))

# Select numeric columns
numeric_cols = features_df.select_dtypes(include=[np.number]).columns

# Remove columns with zero variance
var = features_df[numeric_cols].var()
numeric_cols = var[var > 0].index.tolist()

# Calculate correlation matrix
corr = features_df[numeric_cols].corr()

# Create mask for upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Generate heatmap
sns.heatmap(corr, mask=mask, cmap='coolwarm', annot=False, 
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()

# ## 5. Feature Selection and Engineering
# 
# Let's select and engineer features for our ML model.

# Filter out low-variance features and high correlation features
def select_features(features_df, variance_threshold=0.01, correlation_threshold=0.85):
    """
    Select features by removing low variance and highly correlated features.
    
    Args:
        features_df: DataFrame of features
        variance_threshold: Threshold for variance filtering
        correlation_threshold: Threshold for correlation filtering
    
    Returns:
        Selected feature names
    """
    # Select numeric columns
    numeric_cols = features_df.select_dtypes(include=[np.number]).columns
    
    # Remove file_name column if present
    if 'file_name' in numeric_cols:
        numeric_cols = numeric_cols.drop('file_name')
    
    # Get numeric features
    numeric_features = features_df[numeric_cols]
    
    # Remove low variance features
    var = numeric_features.var()
    high_var_cols = var[var > variance_threshold].index.tolist()
    
    # Remove highly correlated features
    high_var_features = numeric_features[high_var_cols]
    corr = high_var_features.corr().abs()
    
    # Create upper triangle mask
    upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    
    # Find features with correlation greater than threshold
    to_drop = [column for column in upper.columns if any(upper[column] > correlation_threshold)]
    
    # Final selected features
    selected = [col for col in high_var_cols if col not in to_drop]
    
    print(f"Selected {len(selected)} features out of {len(numeric_cols)} total features.")
    return selected

# Get selected features
selected_features = select_features(features_df)
print("Selected features:")
for f in selected_features:
    print(f" - {f}")

# Create final feature matrix
X = features_df[selected_features].values

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ## 6. Unsupervised Learning for Pattern Detection
# 
# Now let's apply unsupervised learning techniques to find patterns that might indicate cognitive decline.

# ### 6.1 Dimensionality Reduction

# Apply PCA for visualization and noise reduction
pca = PCA(n_components=min(len(selected_features), 3))
X_pca = pca.fit_transform(X_scaled)

# Explained variance
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Explained variance ratio: {explained_variance_ratio}")
print(f"Total explained variance: {np.sum(explained_variance_ratio):.4f}")

# Visualize PCA components
plt.figure(figsize=(12, 8))

if X_pca.shape[1] >= 2:  # If we have at least 2 components
    plt.scatter(X_pca[:, 0], X_pca[:, 1], s=100)
    
    # Add file names as labels
    for i, file_name in enumerate(features_df['file_name']):
        plt.annotate(file_name, (X_pca[i, 0], X_pca[i, 1]), fontsize=10)
    
    plt.title('PCA Visualization of Speech Features')
    plt.xlabel(f'PC1 ({explained_variance_ratio[0]:.2f}%)')
    plt.ylabel(f'PC2 ({explained_variance_ratio[1]:.2f}%)')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# ### 6.2 Clustering Analysis

# Apply K-means clustering
k_values = range(2, min(5, len(features_df) + 1))
inertias = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Plot elbow curve to find optimal k
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertias, 'o-')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Select optimal k based on elbow curve or domain knowledge
optimal_k = 2  # This is a placeholder - adjust based on your data
print(f"Selected optimal k = {optimal_k}")

# Perform clustering with optimal k
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Add cluster labels to original dataframe
features_df['cluster'] = clusters

# Visualize clusters in PCA space
plt.figure(figsize=(12, 8))

if X_pca.shape[1] >= 2:  # If we have at least 2 components
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', s=100)
    
    # Add file names as labels
    for i, file_name in enumerate(features_df['file_name']):
        plt.annotate(file_name, (X_pca[i, 0], X_pca[i, 1]), fontsize=10)
    
    plt.title('Clusters in PCA Space')
    plt.xlabel(f'PC1 ({explained_variance_ratio[0]:.2f}%)')
    plt.ylabel(f'PC2 ({explained_variance_ratio[1]:.2f}%)')
    plt.colorbar(scatter, label='Cluster')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# ### 6.3 Anomaly Detection

# Apply Isolation Forest for anomaly detection
iso_forest = IsolationForest(contamination=0.2, random_state=42)
anomaly_scores = iso_forest.fit_predict(X_scaled)

# Convert scores to a more interpretable format (1: normal, -1: anomaly)
# We'll change this to 0: normal, 1: anomaly for easier interpretation
anomalies = [1 if score == -1 else 0 for score in anomaly_scores]

# Add anomaly scores to original dataframe
features_df['anomaly'] = anomalies
features_df['anomaly_score'] = iso_forest.decision_function(X_scaled)
# Convert anomaly score to risk score (higher = higher risk of cognitive decline)
features_df['risk_score'] = 1 - (features_df['anomaly_score'] - features_df['anomaly_score'].min()) / \
                           (features_df['anomaly_score'].max() - features_df['anomaly_score'].min())

# Visualize anomalies in PCA space
plt.figure(figsize=(12, 8))

if X_pca.shape[1] >= 2:  # If we have at least 2 components
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=features_df['risk_score'], cmap='YlOrRd', s=100)
    
    # Add file names as labels
    for i, file_name in enumerate(features_df['file_name']):
        plt.annotate(file_name, (X_pca[i, 0], X_pca[i, 1]), fontsize=10)
    
    plt.title('Cognitive Decline Risk Scores in PCA Space')
    plt.xlabel(f'PC1 ({explained_variance_ratio[0]:.2f}%)')
    plt.ylabel(f'PC2 ({explained_variance_ratio[1]:.2f}%)')
    plt.colorbar(scatter, label='Risk Score')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

# ## 7. Feature Importance Analysis
# 
# Let's analyze which features contribute most to the anomaly detection and clustering.

# ### 7.1 Feature Analysis for Clusters

# Calculate the mean value of each feature for each cluster
cluster_centers = []
cluster_labels = []

for cluster_id in range(optimal_k):
    mask = features_df['cluster'] == cluster_id
    
    if mask.sum() > 0:  # If cluster is not empty
        cluster_mean = features_df.loc[mask, selected_features].mean()
        cluster_centers.append(cluster_mean)
        cluster_labels.append(f"Cluster {cluster_id}")

# Create a dataframe with cluster centers
cluster_centers_df = pd.DataFrame(cluster_centers, index=cluster_labels)

# Feature importance for clusters
plt.figure(figsize=(15, 10))

# Heatmap of cluster centers
sns.heatmap(cluster_centers_df, cmap='RdBu_r', annot=False, linewidths=.5)
plt.title('Feature Values Across Clusters')
plt.tight_layout()
plt.show()

# ### 7.2 Feature Analysis for Anomaly Detection

# Calculate correlation between features and risk score
risk_corr = features_df[selected_features + ['risk_score']].corr()['risk_score'].drop('risk_score')
risk_corr = risk_corr.sort_values(ascending=False)

# Visualize feature importance for risk score
plt.figure(figsize=(12, 8))
risk_corr.plot(kind='bar')
plt.title('Feature Correlation with Risk Score')
plt.xlabel('Features')
plt.ylabel('Correlation')
plt.grid(True, alpha=0.3)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# ## 8. Risk Score Calculation and Report Generation

# Define risk levels based on risk scores
def get_risk_level(score):
    if score < 0.3:
        return "Low"
    elif score < 0.7:
        return "Moderate"
    else:
        return "High"

# Create risk assessment
features_df['risk_level'] = features_df['risk_score'].apply(get_risk_level)

# Sort by risk score (highest first)
risk_assessment = features_df.sort_values('risk_score', ascending=False)

# Calculate feature contribution to risk score
top_features = risk_corr.abs().sort_values(ascending=False).head(5).index.tolist()

# Create a detailed risk assessment
risk_details = []
for _, row in risk_assessment.iterrows():
    file_name = row['file_name']
    risk_score = row['risk_score']
    risk_level = row['risk_level']
    
    # Extract key insights
    key_insights = []
    for feature in top_features:
        if feature in row:
            feature_value = row[feature]
            feature_mean = features_df[feature].mean()
            feature_std = features_df[feature].std()
            
            # Calculate Z-score
            if feature_std > 0:
                z_score = (feature_value - feature_mean) / feature_std
                
                # Add insight if significant deviation
                if abs(z_score) > 1:
                    direction = "higher" if z_score > 0 else "lower"
                    key_insights.append(f"{feature}: {feature_value:.2f} ({direction} than average by {abs(z_score):.1f} std)")
    
    # Add to risk details
    risk_details.append({
        'file_name': file_name,
        'risk_score': risk_score,
        'risk_level': risk_level,
        'key_insights': key_insights
    })

# ### 8.1 Detailed Risk Report

print("\n===== COGNITIVE DECLINE RISK ASSESSMENT =====\n")

for i, details in enumerate(risk_details):
    print(f"Sample {i+1}: {details['file_name']}")
    print(f"Risk Score: {details['risk_score']:.2f}")
    print(f"Risk Level: {details['risk_level']}")
    
    if details['key_insights']:
        print("Key Insights:")
        for insight in details['key_insights']:
            print(f" - {insight}")
    else:
        print("Key Insights: No significant deviations detected")
    
    print("")

# ### 8.2 Risk Score Distribution

plt.figure(figsize=(12, 6))
sns.barplot(x='file_name', y='risk_score', hue='risk_level', data=risk_assessment)
plt.title('Cognitive Decline Risk Scores')
plt.xlabel('Audio Samples')
plt.ylabel('Risk Score')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ### 8.3 Key Feature Visualization for Samples

# Select top features for visualization
top_4_features = risk_corr.abs().sort_values(ascending=False).head(4).index.tolist()

plt.figure(figsize=(15, 12))

for i, feature in enumerate(top_4_features):
    plt.subplot(2, 2, i+1)
    
    # Create an order based on risk score
    order = risk_assessment['file_name'].tolist()
    
    # Plot feature values
    ax = sns.barplot(x='file_name', y=feature, data=risk_assessment, order=order)
    
    # Color bars by risk level
    for j, bar in enumerate(ax.patches):
        risk_level = risk_assessment.iloc[j]['risk_level']
        if risk_level == 'Low':
            bar.set_color('green')
        elif risk_level == 'Moderate':
            bar.set_color('orange')
        else:
            bar.set_color('red')
    
    plt.title(f'{feature}')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ## 9. API-Ready Function for Risk Scoring
# 
# Creating a function that can be deployed as an API endpoint to score new audio files.

def get_cognitive_risk_score(audio_file_path):
    """
    Calculate cognitive decline risk score for a given audio file.
    
    Args:
        audio_file_path: Path to audio file
    
    Returns:
        Dictionary with risk assessment
    """
    try:
        # Extract audio features
        audio_features = extract_audio_features(audio_file_path)
        
        # Transcribe audio
        transcript = transcribe_audio(audio_file_path)
        
        # Extract NLP features
        nlp_features = extract_nlp_features(transcript, os.path.basename(audio_file_path))
        
        # Combine features
        features = {**audio_features, **nlp_features}
        
        # Create a DataFrame with the features
        sample_df = pd.DataFrame([features])
        
        # Select only features used in the model
        selected_cols = [col for col in selected_features if col in sample_df.columns]
        X_sample = sample_df[selected_cols].values
        
        # Handle missing features
        if len(selected_cols) < len(selected_features):
            print(f"Warning: Missing {len(selected_features) - len(selected_cols)} features")
            # Fill missing features with zeros
            X_sample_full = np.zeros((1, len(selected_features)))
            for i, col in enumerate(selected_cols):
                idx = selected_features.index(col)
                X_sample_full[0, idx] = X_sample[0, i]
            X_sample = X_sample_full
        
        # Scale features
        X_sample_scaled = scaler.transform(X_sample)
        
        # Calculate anomaly score
        anomaly_score = iso_forest.decision_function(X_sample_scaled)[0]
        
        # Convert to risk score (higher = higher risk)
        risk_score = 1 - (anomaly_score - features_df['anomaly_score'].min()) / \
                      (features_df['anomaly_score'].max() - features_df['anomaly_score'].min())
        
        # Get risk level
        risk_level = get_risk_level(risk_score)
        
        # Find top contributing features
        feature_contributions = []
        for feature in top_features:
            if feature in sample_df.columns:
                feature_value = sample_df[feature].values[0]
                feature_mean = features_df[feature].mean()
                feature_std = features_df[feature].std()
                
                if feature_std > 0:
                    z_score = (feature_value - feature_mean) / feature_std
                    contribution = abs(z_score) * abs(risk_corr[feature])
                    direction = "higher" if z_score > 0 else "lower"
                    
                    feature_contributions.append({
                        'feature': feature,
                        'value': feature_value,
                        'z_score': z_score,
                        'direction': direction,
                        'contribution': contribution
                    })
        
        # Sort by contribution
        feature_contributions = sorted(feature_contributions, key=lambda x: x['contribution'], reverse=True)
        
        # Prepare result
        result = {
            'file_name': os.path.basename(audio_file_path),
            'risk_score': float(risk_score),
            'risk_level': risk_level,
            'transcript': transcript,
            'key_insights': [
                f"{fc['feature']}: {fc['value']:.2f} ({fc['direction']} than average by {abs(fc['z_score']):.1f} std)"
                for fc in feature_contributions[:3]
            ]
        }
        
        return result
    
    except Exception as e:
        print(f"Error processing file: {e}")
        return {
            'file_name': os.path.basename(audio_file_path),
            'error': str(e),
            'risk_score': None,
            'risk_level': None
        }

# Test the API function on one of our audio files
if audio_files:
    test_file = os.path.join(RAW_AUDIO_DIR, audio_files[0])
    print(f"\nTesting risk score API on: {audio_files[0]}")
    
    result = get_cognitive_risk_score(test_file)
    
    print("\nRisk Assessment:")
    print(f"File: {result['file_name']}")
    print(f"Risk Score: {result['risk_score']:.2f}")
    print(f"Risk Level: {result['risk_level']}")
    print("\nTranscript:")
    print(result['transcript'])
    print("\nKey Insights:")
    for insight in result['key_insights']:
        print(f" - {insight}")

# ## 10. Final Report and Summary
# 
# Generate a final report summarizing the findings and insights.

print("\n===== VOICE-BASED COGNITIVE DECLINE PATTERN DETECTION - FINAL REPORT =====\n")

print("## Overview")
print(f"Analyzed {len(features_df)} voice samples for potential cognitive decline indicators.")
print(f"Found {features_df['anomaly'].sum()} samples with significant deviations from normal patterns.")
print("")

print("## Most Insightful Features")
for i, (feature, corr) in enumerate(risk_corr.abs().sort_values(ascending=False).head(5).items()):
    print(f"{i+1}. {feature} (correlation with risk: {corr:.2f})")
print("")

print("## ML Methods Used")
print("1. Feature Selection: Variance and correlation filtering to identify most relevant features")
print("2. Anomaly Detection: Isolation Forest to identify outlier speech patterns")
print("3. Clustering: K-means to group similar speech patterns")
print("4. Dimensionality Reduction: PCA for visualization and noise reduction")
print("")

print("## Risk Assessment Summary")
for level in ["High", "Moderate", "Low"]:
    count = (features_df['risk_level'] == level).sum()
    print(f"{level} risk: {count} samples")
print("")

print("## Potential Next Steps")
print("1. Validate findings with clinical data from patients with confirmed cognitive conditions")
print("2. Add more cognitive-specific tasks (e.g., word recall, sentence completion)")
print("3. Implement longitudinal tracking to detect changes over time")
print("4. Integrate with other biomarkers for a more comprehensive assessment")
print("5. Apply transfer learning using pre-trained speech models for better feature extraction")
print("")

print("## Conclusion")
print("This proof-of-concept pipeline demonstrates the potential of voice analysis")
print("for detecting cognitive decline patterns. The combination of acoustic and")
print("linguistic features provides a non-invasive way to identify potential warning signs.")
print("Further clinical validation is needed for deployment in real-world scenarios.")

# Save final report to file
with open(os.path.join(DATA_DIR, "cognitive_decline_detection_report.txt"), "w") as f:
    f.write("===== VOICE-BASED COGNITIVE DECLINE PATTERN DETECTION - FINAL REPORT =====\n\n")
    
    f.write("## Overview\n")
    f.write(f"Analyzed {len(features_df)} voice samples for potential cognitive decline indicators.\n")
    f.write(f"Found {features_df['anomaly'].sum()} samples with significant deviations from normal patterns.\n\n")
    
    f.write("## Most Insightful Features\n")
    for i, (feature, corr) in enumerate(risk_corr.abs().sort_values(ascending=False).head(5).items()):
        f.write(f"{i+1}. {feature} (correlation with risk: {corr:.2f})\n")
    f.write("\n")
    
    f.write("## ML Methods Used\n")
    f.write("1. Feature Selection: Variance and correlation filtering to identify most relevant features\n")
    f.write("2. Anomaly Detection: Isolation Forest to identify outlier speech patterns\n")
    f.write("3. Clustering: K-means to group similar speech patterns\n")
    f.write("4. Dimensionality Reduction: PCA for visualization and noise reduction\n\n")
    
    f.write("## Risk Assessment Summary\n")
    for level in ["High", "Moderate", "Low"]:
        count = (features_df['risk_level'] == level).sum()
        f.write(f"{level} risk: {count} samples\n")
    f.write("\n")
    
    f.write("## Potential Next Steps\n")
    f.write("1. Validate findings with clinical data from patients with confirmed cognitive conditions\n")
    f.write("2. Add more cognitive-specific tasks (e.g., word recall, sentence completion)\n")
    f.write("3. Implement longitudinal tracking to detect changes over time\n")
    f.write("4. Integrate with other biomarkers for a more comprehensive assessment\n")
    f.write("5. Apply transfer learning using pre-trained speech models for better feature extraction\n\n")
    
    f.write("## Conclusion\n")
    f.write("This proof-of-concept pipeline demonstrates the potential of voice analysis\n")
    f.write("for detecting cognitive decline patterns. The combination of acoustic and\n")
    f.write("linguistic features provides a non-invasive way to identify potential warning signs.\n")
    f.write("Further clinical validation is needed for deployment in real-world scenarios.")

print(f"\nFinal report saved to: {os.path.join(DATA_DIR, 'cognitive_decline_detection_report.txt')}")
print("\nAnalysis complete!")
```
